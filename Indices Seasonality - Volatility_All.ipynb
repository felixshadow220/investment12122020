```python
import pandas as pd
import numpy as np
import glob
```

```python
def seasonality_function():
    global df
    df=df.drop(['Adj Close', 'Volume'], axis=1)
    df['Date'] = pd.to_datetime(df["Date"], format = '%d/%m/%Y')
    df=df.sort_values(by='Date', ascending=False)
    
    df['Change'] = df['Close']-df['Open']
    df['Absolute Change'] = abs(df['Change'])
    df['Absolute % Change'] = df['Absolute Change']/df['Open']*100
    df['Volatility'] = (df['High'] - df['Low'])/df['Open']*100
    
    numbers = np.arange(0.3, 0.8, 0.1)
    
    percentile = []
    for i in numbers:
        percentile.append(df['Volatility'].quantile(i))
        
    
    month = list(range(1,13))
    index_list = []
    
    for i in month:
        df_month=df[df['Date'].dt.month == i]
        
        frequency_actual = []
        for i in percentile:
            frequency_actual.append(len(df_month[df_month['Volatility']>i]))
        
        frequency_theoretical = []
        for i in numbers:
            frequency_theoretical.append(10-i*10)
        
        index = [i/j*100 for i,j in zip(frequency_actual,frequency_theoretical)]
        
        index_list.append(index)
        
    month_eng = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
    col = np.arange(30, 80, 10)
    Seasonality = pd.DataFrame(index_list, index = month_eng, columns = col).transpose()
    
    df_list.append(Seasonality)
    
    return
    
```
