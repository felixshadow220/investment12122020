```python
import pandas as pd
import numpy as np
import glob
```

```python
def monthly_statistics():
    global df
    df=df.drop(['Adj Close', 'Volume'], axis=1)
    df['Date'] = pd.to_datetime(df["Date"])
    df=df.sort_values(by='Date', ascending=False)
    
    df['Change'] = df['Close']-df['Open']
    df['Absolute Change'] = abs(df['Change'])
    df['Absolute % Change'] = df['Absolute Change']/df['Open']*100
    df['Volatility'] = (df['High'] - df['Low'])/df['Open']*100
    
    numbers = np.arange(0.1, 1.0, 0.1)
    
    percentile = []
    for i in numbers:
        percentile.append(df['Absolute % Change'].quantile(i))
        
    volatility = []
    for i in numbers:
        volatility.append(df['Volatility'].quantile(i))
        
    global list1
    global list2

    list1.append(percentile)
    list2.append(volatility)
    
    return
```

```python
name = list(glob.glob(path))
list_name = []
for i in name:
    list_name.append(i[96:-26])
col = np.arange(10, 100, 10)
```

```python
list1 = []
list2 = []
for fname in glob.glob(path):
    df=pd.read_csv(fname)
    monthly_statistics()
```
