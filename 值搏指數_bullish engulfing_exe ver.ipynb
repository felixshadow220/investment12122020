```python
import pandas as pd
import numpy as np
import glob
```

```python
def index_bullish_engulfing():
    
    #Read and clean data
    global df
    df=df[df['Volume'] != 0]
    df=df.drop(['Adj Close', 'Volume'], axis=1)
    df['Date'] = pd.to_datetime(df["Date"], format='%d/%m/%Y')
    df=df.sort_values(by='Date', ascending=False)
    df['Signal']=''
    
    #Define price signal
    for i in range(5,len(df)-1):
        if df['Open'].iloc[i] < df['Close'].iloc[i] and df['Open'].iloc[i+1] > df['Close'].iloc[i+1] and df['Close'].iloc[i] > df['Open'].iloc[i+1] and df['Open'].iloc[i] < df['Close'].iloc[i+1]:
            df['Signal'].iloc[i] = 'Yes'
        else:
            df['Signal'].iloc[i] = ''
   
    #Define the screening criterion
    minimum_increase = 0.03
    minimum_decrease = -0.03
    
    #Calculate the price changes 1-5 days after the signal appeared 
    df['1 Day After_Up'] = ''
    df['2 Day After_Up'] = ''
    df['3 Day After_Up'] = ''
    df['4 Day After_Up'] = ''
    df['5 Day After_Up'] = ''
    df['1 Day After_Down'] = ''
    df['2 Day After_Down'] = ''
    df['3 Day After_Down'] = ''
    df['4 Day After_Down'] = ''
    df['5 Day After_Down'] = ''
    
    for i in range(1,len(df)):
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-1] - df['Close'].iloc[i])/df['Close'].iloc[i] > minimum_increase:
            df['1 Day After_Up'].iloc[i] = (df['Close'].iloc[i-1] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['1 Day After_Up'].iloc[i] = None
   
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-2] - df['Close'].iloc[i])/df['Close'].iloc[i] > minimum_increase:
            df['2 Day After_Up'].iloc[i] = (df['Close'].iloc[i-2] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['2 Day After_Up'].iloc[i] = None
    
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-3] - df['Close'].iloc[i])/df['Close'].iloc[i] > minimum_increase:
            df['3 Day After_Up'].iloc[i] = (df['Close'].iloc[i-3] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['3 Day After_Up'].iloc[i] = None
    
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-4] - df['Close'].iloc[i])/df['Close'].iloc[i] > minimum_increase:
            df['4 Day After_Up'].iloc[i] = (df['Close'].iloc[i-4] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['4 Day After_Up'].iloc[i] = None
    
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-5] - df['Close'].iloc[i])/df['Close'].iloc[i] > minimum_increase:
            df['5 Day After_Up'].iloc[i] = (df['Close'].iloc[i-5] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['5 Day After_Up'].iloc[i] = None
    
    
    
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-1] - df['Close'].iloc[i])/df['Close'].iloc[i] < minimum_decrease:
            df['1 Day After_Down'].iloc[i] = (df['Close'].iloc[i-1] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['1 Day After_Down'].iloc[i] = None
    
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-2] - df['Close'].iloc[i])/df['Close'].iloc[i] < minimum_decrease:
            df['2 Day After_Down'].iloc[i] = (df['Close'].iloc[i-2] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['2 Day After_Down'].iloc[i] = None
    
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-3] - df['Close'].iloc[i])/df['Close'].iloc[i] < minimum_decrease:
            df['3 Day After_Down'].iloc[i] = (df['Close'].iloc[i-3] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['3 Day After_Down'].iloc[i] = None
   
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-4] - df['Close'].iloc[i])/df['Close'].iloc[i] < minimum_decrease:
            df['4 Day After_Down'].iloc[i] = (df['Close'].iloc[i-4] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['4 Day After_Down'].iloc[i] = None
   
        if df['Signal'].iloc[i] == 'Yes' and (df['Close'].iloc[i-5] - df['Close'].iloc[i])/df['Close'].iloc[i] < minimum_decrease:
            df['5 Day After_Down'].iloc[i] = (df['Close'].iloc[i-5] - df['Close'].iloc[i])/df['Close'].iloc[i]
        else:
            df['5 Day After_Down'].iloc[i] = None
   
    #Sort out the "Yes" data
    Signal=df[df['Signal'] == 'Yes']
    
    # Create a list for looping 
    column_Up = list(df.columns[6:11])
    column_Down = list(df.columns[11:])
    
    # Mean of Increase for 1-5 days after signal
    Median_Up = []
    for i in column_Up:
        Median_Up.append(Signal[i].median())
   
    # Mean of Decrease for 1-5 days after signal    
    Median_Down = []
    for i in column_Down:
        Median_Down.append(Signal[i].median())
    
    # Total no. of increase after signal 
    Number_Up = []
    for i in column_Up:
        Number_Up.append(Signal[i].count())
    
    # Total no. of decrease after signal
    Number_Down = []
    for i in column_Down:
        Number_Down.append(Signal[i].count())
    
    #Total no. of signal
    Number_Signal = Signal['Signal'].count()
    
    # Ratio of Increase
    Ratio_Up = []
    for x in Number_Up:
        Ratio_Up.append(x/Number_Signal)
    
    #Ratio of Decrease
    Ratio_Down = []
    for y in Number_Down:
        Ratio_Down.append(y/Number_Signal)
    
    #Ratio x Median of Increase
    x_Up = []
    for x,y in zip(Ratio_Up, Median_Up):
        x_Up.append(x*y)
    
    #Ratio x Median of Decrease
    x_Down = []
    for x,y in zip(Ratio_Down, Median_Down):
        x_Down.append(x*y)
    
    #Export the initial result
    Dict_1 = {'Median of Increase': Median_Up, 
       'Median of Decrease':Median_Down, 
       'Number of Increase':Number_Up,
        'Number of Decrease':Number_Up,
         'Ratio of Increase':Ratio_Up,
         'Ratio of Decrease':Ratio_Down,
         'Ratio of Increase x Median of Increase': x_Up,
         'Ratio of Increase x Median of Decrease': x_Down}  
    table = pd.DataFrame.from_dict(Dict_1, orient ='index') 
    
    table.rename(columns = {0:'1 Day', 1:'2 Days', 2:'3 Days', 3:'4 Days', 4:'5 Days'}, inplace = True) 
    
    #Replace null result with zero for calculation
    table = table.fillna(0)
    
    # Summing the products across different days for index calculation
    sum_result = table.sum(axis = 1)
    
    #Final result 
    Index_speculation = (sum_result[6] - abs(sum_result[7]))*10000/5
    Index_volatility = (sum_result[6] + abs(sum_result[7]))*10000/5
    
    #Exporting the index out for the for loop
    global index_list1
    global index_list2

    index_list1.append(Index_speculation)
    index_list2.append(Index_volatility)
    
    global occur
    
    occur.append(Number_Signal)
    
    return 
```
